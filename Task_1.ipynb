```python
# LOAD DATA
mc = pd.read_csv('marketing_campaign.csv')
print(f"Raw shape: {mc.shape}")
print(f"Duplicates: {mc.duplicated().sum()}")
print(f"Nulls:\n{mc.isnull().sum()}\n")
```

```python
# REMOVE DUPLICATES
mc = mc.drop_duplicates()
```

```python
# CLEAN COLUMN NAMES
mc.columns = [c.lower().replace(' ', '_') for c in mc.columns]
```

```python
# STANDARDIZE TEXT FIELDS
# GENDER
gender_map = {
    'female': 'Female', 'male': 'Male',
    'f': 'Female', 'm': 'Male',
    'FEMALE': 'Female', 'MALE': 'Male'
}
# mc['gender'] = mc['gender'].map(lambda x: gender_map.get(x.strip(), x.strip().title()))
# The 'gender' column does not exist in the DataFrame, so this line has been commented out to prevent a KeyError.
```

```python
# Country
country_map = {
    'USA': 'United States', 'US': 'United States',
    'UK': 'United Kingdom', 'canada': 'Canada',
    'INDIA': 'India'
}
# mc['country'] = mc['country'].replace(country_map)
```

```python
# Campaign
# mc['campaign'] = mc['campaign'].str.strip().str.title()
# The 'campaign' column does not exist in the DataFrame, so this line has been commented out to prevent a KeyError.
```

```python
# STANDARDIZE CONVERSION
# The 'conversion' column does not exist. Assuming 'response' is the intended column for conversion.
# The 'response' column contains 0s and 1s, which can be directly converted to boolean.
mc['response'] = mc['response'].astype(bool)
```

```python
import pandas as pd
from datetime import datetime
def parse_date(val):
    for fmt in ['%Y-%m-%d', '%d/%m/%Y', '%Y/%m/%d', '%B %d %Y', '%d-%m-%Y']:
        try:
            return datetime.strptime(str(val).strip(), fmt)
        except:
            continue
    return pd.NaT
# Corrected column name from 'date_joined' to 'dt_customer'
mc['dt_customer'] = mc['dt_customer'].apply(parse_date)
```

```python
# HANDLE MISSING NUMERIC VALUES

# Calculate 'age' from 'year_birth'
# Assuming datetime is already imported from a previous cell
from datetime import datetime
current_year = datetime.now().year
mc['age'] = current_year - mc['year_birth']

# Calculate 'total_purchase_amount' from existing 'mnt' columns
mnt_columns = [col for col in mc.columns if col.startswith('mnt')]
mc['total_purchase_amount'] = mc[mnt_columns].sum(axis=1)

# Handle missing values for 'income'
mc['income'] = mc['income'].fillna(mc['income'].median())
```

```python
# EXPORT
mc.to_csv("marketing_campaign_cleaned.csv", index=False)
print(f"✅ Cleaned shape: {mc.shape}")
print(f"Nulls remaining: {mc.isnull().sum().sum()}")
print("\nCleaned dataset saved to marketing_campaign_cleaned.csv")
```
